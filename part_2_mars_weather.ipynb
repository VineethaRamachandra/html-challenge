```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
html_soup = soup(html, 'html.parser')
html_soup 
```

```python
# First determine the HTML table

table = html_soup.find('table', class_= 'table')

# Determine row header
row_header = table.find_all('th')
```

```python
# Extract all rows of data

data_row = table.find_all('tr', class_='data-row')
```

```python
# Determine if columns have the correct headings and data types

row_heading = html_soup.find_all('th')
heading=[]
for header in row_heading:
    title=header.text
    heading.append(title)
    
heading
```

```python
# Get the headings into a dataframe

rows_df = pd.DataFrame(columns=heading)
rows_df
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list

# Loop through the scraped data to create a list of rows
# Create a Pandas DataFrame by using the list of rows and a list of the column names
for row in html_soup.find_all('tr',class_='data-row'):
        data = row.find_all('td')
        row_data=[td.text.strip() for td in data]
        mars=len(rows_df)
        rows_df.loc[mars]=row_data
```

```python
# Confirm DataFrame was created successfully
rows_df
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
rows_df.dtypes
rows_df.info()
```

```python
# Change data types for data analysis
rows_df['id']=rows_df['id'].astype(int)
rows_df['terrestrial_date']=pd.to_datetime(rows_df['terrestrial_date'])
rows_df['sol']=rows_df['sol'].astype(int)
rows_df['ls']=rows_df['ls'].astype(int)
rows_df['month']=rows_df['month'].astype(int)
rows_df['min_temp']=rows_df['min_temp'].astype(float)
rows_df['pressure']=rows_df['pressure'].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
rows_df.info()
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
months_in_mars = rows_df.groupby('month')
months_in_mars['id'].count()
```

```python
# Calculate and print number of months that exist on Mars

total_months_mars=len(months_in_mars['id'].count())
print(f'The number of months that exist on Mars: {total_months_mars}')
```

```python
# 2. How many Martian days' worth of data are there?

maxdata=rows_df['sol'].min()
mindata = rows_df['sol'].max()
unique_date=rows_df['sol'].nunique()

print(f'There are {unique_date} Martian days worth of data in the scraped dataset.')
```

```python
# 3. What is the average low temperature by month?
avg_low_temp = rows_df.groupby(['month']).mean(numeric_only=False)['min_temp']

avg_low_temp_df=pd.DataFrame(avg_low_temp)
avg_low_temp
```

```python
#Export list to JSON file

avg_min_temp_list=[]

for i in range(1,total_months_mars+1):
    weather_per_month={'month_num':"","avg_min_temp":""}
    avg_min_temp_month=rows_df.loc[rows_df['month']==i]['min_temp'].mean()
    round_avg_min_temp_month = round(avg_min_temp_month,0)
    weather_per_month['month_num'] = i
    weather_per_month["avg_min_temp"] = round_avg_min_temp_month
    avg_min_temp_list.append(weather_per_month)    

print(avg_min_temp_list) 
```

```python
json_avg_low_temp = json.dumps(avg_min_temp_list)
# Save json as mars_df_month_temp_df file
rows_df_month_temp=pd.read_json(json_avg_low_temp)
rows_df_month_temp.to_json('mars_df_month_temp.json')
```

```python
# Plot the average temperature by month
avg_low_temp.plot.bar()
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.title('Average temperature by month')
plt.savefig('avg_low_temp.png')
```

```python
# Identify the coldest and hottest months in Curiosity's location
coldes_hottest_temp = avg_low_temp_df.sort_values(by=['min_temp'], ascending=False)
coldes_hottest_temp.plot.bar()
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.title('Coldest and Hottest months in location')
plt.savefig('cold_hottest_temp.png')
```

```python
print(f'From the above plot, we can conclude that, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest.')
```

```python
# 4. Average pressure by Martian month
avg_pres =rows_df.groupby(['month']).mean(numeric_only=False)['pressure']
average_pressure=pd.DataFrame(avg_pres)
average_pressure
```

```python
# Plot the average pressure by month
avg_pressure = avg_pres.sort_values()
avg_pressure.plot.bar()
plt.xlabel('Month')
plt.ylabel('Pressure')
plt.title('Average pressure by Martian month')
plt.savefig('average_pressure.png')
```

```python
print(f' From the above plot, we can conclude that atmospheric pressure is lowest in the sixth month and highest in the ninth month.')
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
days_in_mars = rows_df['terrestrial_date'].loc[(rows_df['ls']==rows_df['ls'][0]) & (rows_df.index!=0)].iloc[0] - rows_df['terrestrial_date'][0]


print(f'One Martial year is equivalent to {days_in_mars.days+1} earth/terrestrial days')
```

```python
#Visually estimate the result by plotting the daily minimum temperature.

rows_df['min_temp'].plot(
    kind='line', 
    xlabel='Number of Terrestrial Days', 
    ylabel='Minimum Temperature')
plt.title('Visualization of daily minimum temperature')
plt.show()

plt.savefig('daily_min_temp.png')
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
